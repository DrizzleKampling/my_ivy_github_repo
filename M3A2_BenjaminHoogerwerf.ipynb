```python
# Sort an arry of 0s, 1s, and 2s
class Solution:
    def sort012(self, arr):
        for i in arr:
            if arr[i] == 0:
                arr.pop(i)
                arr.insert(0, 0)
            elif arr[i] == 2:
                arr.pop(i)
                arr.append(2)
        print(arr)
        
test = Solution()
test.sort012([0, 1, 2, 0, 1, 2])
```

```python
# Binary Search
class Solution:
    def binarysearch(self, arr, k):
        if arr.index(k) == False:
            return -1
        else:
            return arr.index(k)

test = Solution()
test.binarysearch([1, 2, 3, 4, 5], 4)
```
