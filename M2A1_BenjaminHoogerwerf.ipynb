```python
# 4.1
secret = 5
guess = 4

if guess < secret:
    print("too low")
elif guess > secret:
    print("too high")
elif guess == secret:
    print("just right!")
```

```python
# 4.2
small = True
green = False
food = ""

if (small == True) and (green == False):
    food = "cherry"
elif (small == True) and (green == True):
    food = "pea"
elif (small == False) and (green == False):
    food = "pumpkin"
elif (small == False) and (green == True):
    food = "watermelon"

print(food)
```

```python
# 6.2
guess_me = 7
number = 1

while number != guess_me:
    if number < guess_me:
        print("too low")
    elif number > guess_me:
        print("oops")
        break
    number += 1

if number == guess_me:
    print("found it!")
        
```

```python
# 6.3
guess_me = 5

for number in range(10):
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    elif number > guess_me:
        print("oops")
        break
```
