Dividing the variable "seconds_per_day" by the variable "seconds_per_hour" with floating point division first, then by integer division, returned two different results. The reason this happens is because floating point division represents any decimals that result from the division and, if there are none, simply appends ".0" to the end of the number for clarity. By contrast, integer division cuts out any decimal numbers entirely, not even bothering to round to the nearest whole number. There
